Plotting sin linea. Para trazar solo los marecadores , puedes usar notaciuon de cadena de acceso directo "o", que significa "anillos" Ejemplo:Dibuja dos puntos
en el diagrama, uno en la posicion (1,3) y otro en la posicion (8,10)

```python
import  matplotlib.pyplot as plt
import numpy as np

puntosx=np.array([1,8])
puntosy=np.array([3,10])

plt.plot(puntosx,puntosy,'o')
plt.show()
```

Diagrama de dispersion (scatter)
-
Se emplean para mostrar la relacion entre dos variables numericas ,
detectando correlaciones o agrupamientops , ejemplo:


```python
import matplotlib.pyplot as plt
altura=[150,160,170,180,190,200]
peso=[50,60,65,80,85,95]

plt.scatter(altura, peso, color="red",marker="x")
plt.title("relacion entre altura y el peso")
plt.xlabel("altura(cm)")
plt.ylabel("peso(kg)")
plt.show()
```

graficos de barra(bar) sirven para comprobar categorias o valoriables tanto en orientacion vertical como horizontal
ejemplo de barras verticales:


```python
import matplotlib.pyplot as plt
categorias =["A","B","C","D"]
valores=[4,7,1,8]
plt.bar(categorias,valores,color="orange")
plt.title("Grafica de barras")
plt.xlabel("categorias")
plt.ylabel("valores")
plt.show()
```

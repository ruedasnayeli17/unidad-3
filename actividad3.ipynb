Conceptos basicos de la libreria matplotlib
--
Es una libreria de trazado de grafico de bajo nivel
en python que sirve como una herramienta de visualizacion
_matplotlib esta escrito principalmente en python con algunos segmentos escritos en C
-
Instalacion de matplotlib
Si tienes python y pip ya instalado en tu sistema , instala matplotlib usando  el siguiente comando:
--


```python
pip install matplotlib
```

Importar matplotlib
--
una vez instalado matplotlib importalo en tus aplicaciones agregando el modulo import
--
ejemplo:
--


```python
import matplotlib
```

Matplotlib pyplot
Pyplot se encuentra dentro del submodulo pyplot y normalmente se importan bajo el alias plt

```python
import matplotlib.pyplot as plt
```

Ahora el paquete pyplot puede denominarse plt
